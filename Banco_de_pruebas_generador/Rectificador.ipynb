```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df_voc= pd.read_csv('Datos_voc_6-6-25_DC.csv')
df_vload= pd.read_csv('V_load_DC_6-6-25.csv')
```

```python
#Datos circuito abierto
plt.figure(figsize=(8, 5))
plt.plot(df_voc['freq 1'], df_voc['Voc 1'], label='V_oc', color='black')#, s=10)
#plt.scatter(df_voc['freq 2'], df_voc['Voc 2'], label='V_oc', color='red',s=10)
#plt.scatter(df_voc['freq 3'], df_voc['Voc 3'], label='V_oc', color='red',s=10)
plt.title(r'$V_{oc}$ vs Frecuencia circuito rectificador en paralelo')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel(r'$V_{oc}$ [V]')
plt.show()

plt.figure(figsize=(8, 5))
plt.scatter(df_voc['ms @ 12V'], df_voc['freq 1'], label='V_oc', color='black', s=10)
plt.scatter(df_voc['ms @ 12V'], df_voc['freq 2'], label='V_oc', color='red',s=10)
plt.scatter(df_voc['ms @ 12V'], df_voc['freq 3'], label='V_oc', color='red',s=10)
plt.title(r'$V_{oc}$ vs Frecuencia circuito rectificador en paralelo')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel(r'$V_{oc}$ [V]')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df_vload['freq 1'], df_vload['Vload 1'], label='V_oc', color='black', s=10)
plt.scatter(df_vload['freq 2'], df_vload['Vload 2'], label='V_oc', color='red',s=10)
plt.scatter(df_vload['freq 3'], df_vload['Vload 3'], label='V_oc', color='red',s=10)
plt.title(r'$V_{L}$ vs Frecuencia circuito rectificador en paralelo')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel(r'$V_{L}$ [V]')
plt.show()

plt.figure(figsize=(8, 5))
plt.scatter(df_vload['Ohm @ 2ms-12V'], df_vload['Vload 1'], label='V_oc', color='black', s=10)
plt.scatter(df_vload['Ohm @ 2ms-12V'], df_vload['Vload 2'], label='V_oc', color='red',s=10)
plt.scatter(df_vload['Ohm @ 2ms-12V'], df_vload['Vload 3'], label='V_oc', color='red',s=10)
plt.title(r'$V_{L}$ vs Resistencia de carga circuito rectificador en paralelo')
plt.xlabel('Resistencia de carga [Ohm]')
plt.ylabel(r'$V_{L}$ [V]')
plt.show()

df_vload['Pl 1'] = df_vload['Vload 1'] **2/ df_vload['Ohm @ 2ms-12V']
df_vload['Pl 2'] = df_vload['Vload 2'] **2/ df_vload['Ohm @ 2ms-12V']
df_vload['Pl 3'] = df_vload['Vload 3'] **2/ df_vload['Ohm @ 2ms-12V']

plt.figure(figsize=(8, 5))
plt.scatter(df_vload['Ohm @ 2ms-12V'], df_vload['Pl 1'], label='P_load', color='black', s=10)
plt.scatter(df_vload['Ohm @ 2ms-12V'], df_vload['Pl 2'], label='P_load', color='red',s=10)
plt.scatter(df_vload['Ohm @ 2ms-12V'], df_vload['Pl 3'], label='P_load', color='red',s=10)
plt.title(r'$P_{L}$ vs Resistencia de carga circuito rectificador en paralelo')
plt.xlabel('Resistencia de carga [Ohm]')
plt.ylim(0,4)
plt.ylabel(r'$P_{L}$ [W]')
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(df_vload['freq 1'], df_vload['Pl 1'], label='P_load', color='black')#, s=10)
#plt.scatter(df_vload['freq 2'], df_vload['Pl 2'], label='P_load', color='red',s=10)
#plt.scatter(df_vload['freq 3'], df_vload['Pl 3'], label='P_load', color='red',s=10)
plt.title(r'$P_{L}$ vs Frecuencia circuito rectificador en paralelo')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel(r'$P_{L}$ [W]')
plt.show()

df_vload['I 1'] = df_vload['Vload 1'] / df_vload['Ohm @ 2ms-12V']
df_vload['I 2'] = df_vload['Vload 2'] / df_vload['Ohm @ 2ms-12V']
df_vload['I 3'] = df_vload['Vload 3'] / df_vload['Ohm @ 2ms-12V']

plt.figure(figsize=(8, 5))
plt.plot(df_vload['Vload 1'], df_vload['I 1'], label='I_load', color='red')#, s=10)
#plt.scatter(df_vload['Vload 2'], df_vload['I 2'], label='I_load', color='red',s=10)
#plt.scatter(df_vload['Vload 3'], df_vload['I 3'], label='I_load', color='red',s=10)
plt.title(r'$I_{L}$ vs $V_{L}$ circuito rectificador en paralelo')
plt.xlabel(r'$V_{L}$ [V]')
plt.ylabel(r'$I_{L}$ [A]')
plt.legend(loc='upper left')
#plt.ylim(0, max(df_vload['I 1']) + 0.05)
#double y-axis
plt.twinx()
plt.plot(df_vload['Vload 1'], df_vload['Pl 1'], label='P_load', color='black')#, s=10)
#plt.scatter(df_vload['Vload 2'], df_vload['Pl 2'], label='P_load', color='red',s=10)
#plt.scatter(df_vload['Vload 3'], df_vload['Pl 3'], label='P_load', color='red',s=10)
plt.ylabel(r'$P_{L}$ [W]')
plt.title(r'$I_{L}$ vs $V_{L}$ circuito rectificador en paralelo')
plt.legend(loc='center left')
plt.xlim(0, max(df_vload['Vload 1']) + 0.5)
plt.ylim(0, max(df_vload['Pl 1']) + 0.5)
plt.grid()
plt.show()
```
